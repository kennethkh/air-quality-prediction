This shows that there are no missing values in the dataset. But, as described in the UCI repository documentation, the actual missing values are tagged with -200 values.

```python
# counting the number of times -200 appears in the data
air_quality_data.isin([-200]).sum(axis=0)
```

Handling the missing values:

```python
# convert all -200 values into NaN values
air_quality_data = air_quality_data.replace(to_replace = -200, value = np.nan)
```

```python
air_quality_data.isnull().sum()
```

This shows the actual number of missing values

Now, we will fill in the missing values with the means of their respective columns.

```python
column_means = air_quality_data.mean(numeric_only=True)
```

```python
air_quality_data = air_quality_data.fillna(column_means)
```

```python
air_quality_data.tail()
```

```python
air_quality_data.isnull().sum()
```

Two tasks can be performed on this dataset:


*   Forecasting
*   Regression



Forecasting with FB Prophet Algorithm:

```python
# converting the date data from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
# converting the time data from HH.MM.SS to HH:MM:SS
time_info = air_quality_data['Time']
time_info = time_info.apply(lambda x: x.replace('.',':'))
print(time_info)
```

```python
# combining the two series date_info and time_info
date_time = pd.concat([date_info, time_info], axis = 1)
date_time.head()
```

```python
# converting the Date and Time columns of date_time dataframe into strings and combining them into a single column
date_time ['DateTime'] = date_time['Date'].astype(str)+ ' ' + date_time['Time'].astype(str)
date_time.head()
```

```python
date_time.info()
```

Now, we have to convert DateTime into datetime datatype.

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['DateTime'])
data.head()
```

```python
data['y'] = air_quality_data['RH']
data.head()
```

```python
! pip install prophet
```

```python
from prophet import Prophet
```

```python
# training the Prophet model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='h')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
